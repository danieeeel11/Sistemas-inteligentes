#**1) Abstract**

El algoritmo de colonia de hormigas (ACO) es una estrategia de optimizacion inspirada en el comportamiento de las hormigas que ha demostrado eficacia en la resolucion de problemas combinatorios. A través de la deposición y seguimiento de feromonas, las "hormigas artificiales” construyen soluciones potenciales, actualizando dinamicamente las feromonas para guiar la exploración del espacio de búsqueda. Aplicado en la optimizacióon de rutas para vehíıculos autónomos en entornos urbanos, el ACO ha destacado por su capacidad para reducir tiempos de entrega y costos operativos.

#**2) Introducción**


La implementación y comparación de algoritmos de búsqueda como A* y Dijkstra junto con el algoritmo de colonia de hormigas (ACO) constituye un enfoque crucial para evaluar la eficacia computacional en la optimización de rutas de vehículos autónomos de entrega en la ciudad de Bogotá. Mientras que A* y Dijkstra son reconocidos por su capacidad para encontrar caminos más cortos en grafos ponderados, ACO se inspira en el comportamiento colectivo de las colonias de hormigas para abordar problemas de optimización combinatoria.

Esta investigación se propone comparar el rendimiento de estos enfoques en términos de tiempo de ejecución y calidad de las rutas generadas, así como la distancia recorrida, el npumero de iteraciones del algoritmo y el tiempo empleado aproximado en encuntrar la solución. Al considerar factores como la velocidad y direccionalidad de las calles, carreras y avenidas, se busca optimizar el tiempo y la distancia empleados en la entrega de mercancías. La implementación y evaluación de estos algoritmos ofrecerá información valiosa para mejorar la eficiencia logística y el abastecimiento en la dinámica urbe de Bogotá.

#**2) Fundamentos del Algoritmo de Colonia de Hormigas**

El algoritmo de la colonia de hormigas se basa en el comportamiento natural de las hormigas para encontrar la mejor ruta entre un punto de inicio (colonia) y una fuente de alimento (comida). En el contexto del proyecto de optimización de rutas de entrega en Bogotá, el nodo inicial se consideraría el centro de abastecimiento principal y los nodos de entrega serían los destinos de los paquetes.

El funcionamiento del algoritmo implica que las hormigas (representadas como agentes virtuales) exploran diferentes caminos desde el centro de abastecimiento hacia los nodos de entrega. Durante esta exploración, las hormigas depositan feromonas en los caminos recorridos. La cantidad de feromona depositada es proporcional a la calidad del camino (en este caso, la eficiencia en términos de tiempo y distancia).

A medida que más hormigas recorren un camino específico, la concentración de feromonas en ese camino aumenta, lo que lo hace más atractivo para las siguientes hormigas. Este proceso se repite iterativamente, permitiendo que las mejores rutas acumulen más feromonas y, por lo tanto, se conviertan en las preferidas por las hormigas.

Al final de múltiples iteraciones, las rutas con mayor concentración de feromonas representan las mejores opciones de entrega de paquetes, ya que indican los caminos más eficientes en términos de tiempo y distancia. En resumen, el algoritmo de la colonia de hormigas encuentra la mejor ruta entre el centro de abastecimiento y los nodos de entrega mediante la simulación del comportamiento de las hormigas y la aplicación de principios de optimización basados en feromonas.

#**3) Metodología de Implementacion del Algoritmo en la Solución del Problema**

En este contexto, cada ubicación de entrega (destino) y los puntos clave de la ciudad (intersecciones, centros logísticos, etc.) son considerados como nodos en un grafo. Las conexiones entre los nodos representan las rutas posibles que los vehículos pueden tomar, con una distancia (o tiempo de viaje estimado) y posiblemente otros factores como el tráfico actual asociados a cada arista.

Cada arista en el grafo tiene asociado un nivel de feromonas que indica la calidad de esa ruta en términos de eficiencia y confiabilidad. Estos niveles de feromona pueden ser inicializados con un valor arbitrario o calculados dinámicamente en función de la experiencia pasada de los vehículos en esas rutas. Además, se considera la capacidad limitada de los vehículos para llevar paquetes al planificar las rutas, asegurándose de que no sobrepasen su capacidad durante el viaje.

El algoritmo de las hormigas guía a los vehículos autónomos en la selección de rutas óptimas. Las hormigas (o vehículos) construyen rutas visitando nodos y seleccionando aristas basadas en las feromonas depositadas y en heurísticas locales, como la distancia restante al destino o la capacidad actual del vehículo. Después de que todas las hormigas hayan construido sus rutas, se actualizan los niveles de feromonas en las aristas en función de la calidad de las soluciones encontradas.

La actualización de las feromonas sigue una regla de "evaporación-deposición", donde se considera la tasa de evaporación de la feromona y la cantidad de feromona depositada por cada hormiga. El proceso de búsqueda se repite hasta que se cumple un criterio de terminación, como un número máximo de iteraciones o la convergencia a una solución aceptable.

Finalmente, las rutas obtenidas por los vehículos autónomos se evalúan en función de criterios como la eficiencia en la entrega de paquetes, el tiempo total de viaje y la capacidad utilizada de los vehículos. Estos resultados se comparan con otras técnicas de optimización o con soluciones manuales para validar la efectividad del enfoque propuesto.

##5.2) Importación y limpieza del mapa

```python
# Se define el nombre del lugar de interés, en este caso, "Bogotá, Colombia".
place_name = "Bogota, Colombia"

# Se establece la distancia de búfer, que determina el área cubierta alrededor del lugar de interés.
buffer_dist = 10000  # metros (aumentar este valor para aumentar el área cubierta y obtener más nodos)

# Se utiliza la función graph_from_place de OSMnx para generar un grafo de carreteras aptas para conducir a partir del lugar de interés.
G = ox.graph_from_place(place_name, network_type="drive", buffer_dist=buffer_dist)
```

```python
for edge in G.edges:
    # Limpiar el atributo "maxspeed", algunos valores son listas, algunos son cadenas, algunos son None
    maxspeed = 50
    if "maxspeed" in G.edges[edge]:
        maxspeed_value = G.edges[edge]["maxspeed"]
        if type(maxspeed_value) == list:
            # Extraer los valores numéricos de la lista y encontrar el mínimo
            speeds = [int(re.search(r'\d+', speed).group()) for speed in maxspeed_value if re.search(r'\d+', speed)]
            maxspeed = min(speeds)
        elif type(maxspeed_value) == str:
            # Extraer el valor numérico de la cadena, ignorando la unidad
            parte_numerica = re.search(r'\d+', maxspeed_value)
            if parte_numerica:
                maxspeed = int(parte_numerica.group())
    G.edges[edge]["maxspeed"] = maxspeed
    # Añadir el atributo "weight" (tiempo = distancia / velocidad)
    G.edges[edge]["weight"] = G.edges[edge]["length"] / maxspeed
```

##5.3) Ayudas visuales del mapa

```python
def style_unvisited_edge(edge):
    """
    Función para establecer el estilo de una arista no visitada.

    Parámetros:
    - edge: Arista a la que se aplicará el estilo.

    Salida:
    - No hay salida explícita, ya que la función modifica las propiedades de la arista en el grafo G globalmente definido.
    """
    # Establecer color, transparencia y grosor de línea para la arista no visitada
    G.edges[edge]["color"] = "#d36206"
    G.edges[edge]["alpha"] = 0.2
    G.edges[edge]["linewidth"] = 0.5

def style_visited_edge(edge):
    """
    Función para establecer el estilo de una arista visitada.

    Parámetros:
    - edge: Arista a la que se aplicará el estilo.

    Salida:
    - No hay salida explícita, ya que la función modifica las propiedades de la arista en el grafo G globalmente definido.
    """
    # Establecer color, transparencia y grosor de línea para la arista visitada
    G.edges[edge]["color"] = "#d36206"
    G.edges[edge]["alpha"] = 1
    G.edges[edge]["linewidth"] = 1

def style_active_edge(edge):
    """
    Función para establecer el estilo de una arista activa.

    Parámetros:
    - edge: Arista a la que se aplicará el estilo.

    Salida:
    - No hay salida explícita, ya que la función modifica las propiedades de la arista en el grafo G globalmente definido.
    """
    # Establecer color, transparencia y grosor de línea para la arista activa
    G.edges[edge]["color"] = '#e8a900'
    G.edges[edge]["alpha"] = 1
    G.edges[edge]["linewidth"] = 1

def style_path_edge(edge):
    """
    Función para establecer el estilo de una arista de camino.

    Parámetros:
    - edge: Arista a la que se aplicará el estilo.

    Salida:
    - No hay salida explícita, ya que la función modifica las propiedades de la arista en el grafo G globalmente definido.
    """
    # Establecer color, transparencia y grosor de línea para la arista de camino
    G.edges[edge]["color"] = "white"
    G.edges[edge]["alpha"] = 1
    G.edges[edge]["linewidth"] = 1
```

```python
def plot_graph():
    """
    Función para graficar un grafo utilizando la librería OSMnx.

    Parámetros:
    - No hay parámetros de entrada explícitos, ya que la función utiliza el grafo G definido globalmente.

    Salida:
    - Gráfico visualizado en la interfaz gráfica.
    """
    # Importar la función plot_graph de la librería ox (OpenStreetMap)
    ox.plot_graph(
        # Graficar el grafo G
        G,
        # Tamaño de los nodos obtenido de la propiedad "size" de cada nodo en G
        node_size =  [ G.nodes[node]["size"] for node in G.nodes ],
        # Color de las aristas obtenido de la propiedad "color" de cada arista en G
        edge_color = [ G.edges[edge]["color"] for edge in G.edges ],
        # Transparencia de las aristas obtenida de la propiedad "alpha" de cada arista en G
        edge_alpha = [ G.edges[edge]["alpha"] for edge in G.edges ],
        # Grosor de las aristas obtenido de la propiedad "linewidth" de cada arista en G
        edge_linewidth = [ G.edges[edge]["linewidth"] for edge in G.edges ],
        # Color de los nodos establecido en blanco
        node_color = "white",
        # Color de fondo establecido en un tono oscuro de rojo
        bgcolor = "#18080e"
    )
```

```python
def plot_heatmap(algorithm):
    """
    Función para graficar un mapa de calor de un algoritmo específico en un grafo.

    Parámetros:
    - algorithm: Nombre del algoritmo para el cual se generará el mapa de calor.

    Salida:
    - Gráfico visualizado en la interfaz gráfica.
    """
    # Obtener los colores de las aristas según la atributo específico del algoritmo
    edge_colors = ox.plot.get_edge_colors_by_attr(G, f"{algorithm}_uses", cmap="hot")
    # Graficar el grafo G con los colores de las aristas obtenidos
    fig, _ = ox.plot_graph(
        G,
        # No mostrar nodos en el gráfico de calor
        node_size = 0,
        # Utilizar los colores de las aristas obtenidos
        edge_color = edge_colors,
        # Color de fondo establecido en un tono oscuro de rojo
        bgcolor = "#18080e"
    )
```

## 5.4) Implementación de los algoritmos

```python
def dijkstra(orig, dest, plot=False):
    """
    Función para realizar el algoritmo de Dijkstra en un grafo dado desde un nodo de origen a un nodo de destino.

    Parámetros:
    - orig: Nodo de origen del camino.
    - dest: Nodo de destino del camino.
    - plot: Booleano que indica si se debe mostrar la visualización del proceso paso a paso. Por defecto es False.

    Salida:
    - Si se especifica plot como True, se mostrará la visualización del proceso paso a paso. Si no, la función solo calculará el camino más corto.
    """
    # Inicializar nodos y aristas del grafo para el algoritmo
    for node in G.nodes:
        G.nodes[node]["visited"] = False
        G.nodes[node]["distance"] = float("inf")
        G.nodes[node]["previous"] = None
        G.nodes[node]["size"] = 0
    for edge in G.edges:
        style_unvisited_edge(edge)

    # Inicializar nodo de origen
    G.nodes[orig]["distance"] = 0
    G.nodes[orig]["size"] = 50
    G.nodes[dest]["size"] = 50

    # Inicializar cola de prioridad para nodos
    pq = [(0, orig)]
    step = 0

    # Ejecutar el algoritmo de Dijkstra
    while pq:
        _, node = heapq.heappop(pq)
        if node == dest:
            if plot:
                print("Iteraciones:", step)
                plot_graph()
            return
        if G.nodes[node]["visited"]:
            continue
        G.nodes[node]["visited"] = True

        # Explorar vecinos del nodo actual
        for edge in G.out_edges(node):
            style_visited_edge((edge[0], edge[1], 0))
            neighbor = edge[1]
            weight = G.edges[(edge[0], edge[1], 0)]["weight"]
            if G.nodes[neighbor]["distance"] > G.nodes[node]["distance"] + weight:
                G.nodes[neighbor]["distance"] = G.nodes[node]["distance"] + weight
                G.nodes[neighbor]["previous"] = node
                heapq.heappush(pq, (G.nodes[neighbor]["distance"], neighbor))
                for edge2 in G.out_edges(neighbor):
                    style_active_edge((edge2[0], edge2[1], 0))
        step += 1
```

```python
def distance(node1, node2):
    """
    Función para calcular la distancia euclidiana entre dos nodos en un grafo.

    Parámetros:
    - node1: Nodo de origen.
    - node2: Nodo de destino.

    Salida:
    - Distancia euclidiana entre los dos nodos.
    """
    x1, y1 = G.nodes[node1]["x"], G.nodes[node1]["y"]
    x2, y2 = G.nodes[node2]["x"], G.nodes[node2]["y"]
    return ((x2 - x1)**2 + (y2 - y1)**2)**0.5

def a_star(orig, dest, plot=False):
    """
    Función para realizar el algoritmo A* en un grafo dado desde un nodo de origen a un nodo de destino.

    Parámetros:
    - orig: Nodo de origen del camino.
    - dest: Nodo de destino del camino.
    - plot: Booleano que indica si se debe mostrar la visualización del proceso paso a paso. Por defecto es False.

    Salida:
    - Si se especifica plot como True, se mostrará la visualización del proceso paso a paso. Si no, la función solo calculará el camino más corto.
    """
    # Inicializar nodos y aristas del grafo para el algoritmo
    for node in G.nodes:
        G.nodes[node]["previous"] = None
        G.nodes[node]["size"] = 0
        G.nodes[node]["g_score"] = float("inf")
        G.nodes[node]["f_score"] = float("inf")
    for edge in G.edges:
        style_unvisited_edge(edge)

    # Inicializar nodo de origen
    G.nodes[orig]["size"] = 50
    G.nodes[dest]["size"] = 50
    G.nodes[orig]["g_score"] = 0
    G.nodes[orig]["f_score"] = distance(orig, dest)
    pq = [(G.nodes[orig]["f_score"], orig)]
    step = 0

    # Ejecutar el algoritmo A*
    while pq:
        _, node = heapq.heappop(pq)
        if node == dest:
            if plot:
                print("Iteraciones:", step)
                plot_graph()
            return
        for edge in G.out_edges(node):
            style_visited_edge((edge[0], edge[1], 0))
            neighbor = edge[1]
            tentative_g_score = G.nodes[node]["g_score"] + distance(node, neighbor)
            if tentative_g_score < G.nodes[neighbor]["g_score"]:
                G.nodes[neighbor]["previous"] = node
                G.nodes[neighbor]["g_score"] = tentative_g_score
                G.nodes[neighbor]["f_score"] = tentative_g_score + distance(neighbor, dest)
                heapq.heappush(pq, (G.nodes[neighbor]["f_score"], neighbor))
                for edge2 in G.out_edges(neighbor):
                    style_active_edge((edge2[0], edge2[1], 0))
        step += 1
```

```python
def reconstruct_path(orig, dest, plot=False, algorithm=None):
    """
    Función para reconstruir el camino más corto desde el nodo de destino al nodo de origen.

    Parámetros:
    - orig: Nodo de origen del camino.
    - dest: Nodo de destino del camino.
    - plot: Booleano que indica si se debe mostrar la visualización del proceso paso a paso. Por defecto es False.
    - algorithm: Nombre del algoritmo utilizado para encontrar el camino. Por defecto es None.

    Salida:
    - Si se especifica plot como True, se mostrará la visualización del proceso paso a paso. Si no, la función solo calculará la distancia del camino y, opcionalmente, actualizará el atributo de uso del algoritmo.
    """
    # Establecer el estilo inicial para todas las aristas del grafo
    for edge in G.edges:
        style_unvisited_edge(edge)

    # Inicializar variables para calcular la distancia y la velocidad
    dist = 0
    speeds = []
    curr = dest

    # Reconstruir el camino desde el destino hasta el origen
    while curr != orig:
        prev = G.nodes[curr]["previous"]
        # Actualizar la distancia total y recopilar velocidades
        dist += G.edges[(prev, curr, 0)]["length"]
        speeds.append(G.edges[(prev, curr, 0)]["maxspeed"])
        # Establecer el estilo de la arista como parte del camino reconstruido
        style_path_edge((prev, curr, 0))
        # Opcionalmente, actualizar el atributo de uso del algoritmo si se proporciona
        if algorithm:
            G.edges[(prev, curr, 0)][f"{algorithm}_uses"] = G.edges[(prev, curr, 0)].get(f"{algorithm}_uses", 0) + 1
        curr = prev

    # Calcular la distancia total en kilómetros
    dist /= 1000

    # Mostrar información adicional si se solicita
    if plot:
        print(f"Distance: {dist:.2f} km")
        print(f"Avg. speed: {sum(speeds)/len(speeds):.2f} km/h")
        print(f"Total time: {round(dist/(sum(speeds)/len(speeds)) * 60, 2)} minutes")
        plot_graph()
```

```python
def ant_colony_optimization(start, end, num_ants=10, evaporation_rate=0.1, alpha=1, beta=2, plot=False):
    pheromone = {edge: 1 for edge in G.edges}  # Initialize pheromone levels on edges
    for _ in range(num_ants):
        ant_path = []
        current_node = start
        while current_node != end:
            unvisited_neighbors = [neighbor for neighbor in G.neighbors(current_node) if neighbor not in ant_path]
            if not unvisited_neighbors:  # If there are no unvisited neighbors, backtrack
                break
            probabilities = []
            total = 0
            for neighbor in unvisited_neighbors:
                pheromone_level = pheromone[(current_node, neighbor, 0)]
                distance = G.edges[(current_node, neighbor, 0)]["length"]
                probabilities.append((neighbor, (pheromone_level ** alpha) * ((1 / distance) ** beta)))
                total += probabilities[-1][1]
            probabilities = [(neighbor, prob / total) for neighbor, prob in probabilities]
            next_node = random.choices([neighbor for neighbor, _ in probabilities], [prob for _, prob in probabilities])[0]
            ant_path.append((current_node, next_node, 0))
            current_node = next_node

        # Update pheromone levels based on the ant's path
        for edge in ant_path:
            pheromone[edge] = (1 - evaporation_rate) * pheromone[edge] + (1 / G.edges[edge]["length"])

    if plot:
        plot_heatmap("ant_colony_optimization")

    # Reconstruct the path from start to end
    reconstruct_path(start, end, plot=plot, algorithm="ant_colony_optimization")

# Experiment 2
start = random.choice(list(G.nodes))
end = random.choice(list(G.nodes))

ant_colony_optimization(start, end, plot=True)
```
